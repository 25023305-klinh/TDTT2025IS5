```python
#bai1
def max_of_two(a:int, b:int)-> int:
    return max(a,b)
try:
    a, b = map(int, input().split())
    print(max_of_two(a, b))
except ValueError:
    print("Nhap lai du lieu")
    
```

```python
#bai3
import math
def prime_number(n: int) -> bool:
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return (n >= 2)
    
n = int(input())
for x in range(2, n + 1):
    if(prime_number(x)):
        print(x, end=" ")
        
```

```python
#bai4
#so hoan hao la so co tong cac uoc bang chinh no
def so_hoan_hao(n: int):
    tong = 0
    for i in range(1,n):
        if n % i == 0:
            tong += i
    if tong == n:
        return True
    else:
        return False 
n = int(input())
print(so_hoan_hao(n))
```

```python
#bai6
def giaithua(n: int):
    gt = 1
    for i in range(1, n + 1):
        gt *= i
    return gt
    
n = int(input())
print(giaithua(n))
```

```python
#bai7
def pheptoan(a, b, toantu):
    if toantu == '+':
        s = a + b
    elif toantu == '-':
        s = a - b
    elif toantu == '*':
        s = a * b
    elif toantu == '/':
        if b == 0:
            return "loi"
        else:
            s = a / b
    else:
        return "toantu kh hop le"
    return round(s, 2)
            

a, b = map(int, input().split())
toantu = input()
print(pheptoan(a, b, toantu))
```

```python
#bai8
def khoangcach(x: int, y: int):
    kc = 0
    for i in range(x+1, y):
        kc += 1
    return kc

x, y = map(int, input().split())
print(khoangcach(x, y))
```

```python
#bai9
def tongcacso(n: int):
    tong = 0
    for i in str(n):
        tong += int(i)
    return tong

n = int(input())
print(tongcacso(n))
```

```python
def dangcau(a, b) -> bool:
    if len(a) != len(b):
        return False
    ab = {}
    ba = {}
    for i in range(len(a)):
        if a[i] in ab:
            if ab[a[i]] != b[i]:
                return False
        else:
            # nếu b[i] đã có ánh xạ nhưng không phải từ a[i] → sai
            if b[i] in ba and ba[b[i]] != a[i]:
                return False
            # luu cap a-b vao neu sau bi trung thi kh ktra lai
            ab[a[i]] = b[i]
            ba[b[i]] = a[i]
    return True

a = input()
b = input()
print(dangcau(a, b))
```
