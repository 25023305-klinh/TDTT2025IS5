```python
#bai1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
n = int(input())
arr = list(map(int, input().split()))
target = int(input())
print(binary_search(arr, 0, n - 1, target ))
```

```python
#bai2 
def demsolan(arr, x):
    dem = 0
    for i in arr:
        if i == x:
            dem += 1
    return dem
arr = list(map(int, input().split()))
x = int(input())
print(demsolan(arr, x))
```

```python
#bai3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
```

```python
#bai3
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_i = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_i]:
                min_i = j
        arr[i], arr[min_i] = arr[min_i], arr[i]
    return arr
arr = list(map(int, input().split()))
print(selection_sort(arr))
```

```python
#bai4
def xhnhieu(arr):
    dic = {}
    lst = []
    for i in arr:
        if i not in lst:
            lst.append(i)
            dic[i] = 1
        else:
            dic[i] += 1
    max_v = max(dic.values())
    key = []
    for k, v in dic.items():
        if v == max_v:
            key.append(k)
    kq = min(key)
    return f"{kq} xuat hien nhieu nhat som nhat, {max_v} lan"
arr = list(map(int, input().split()))
print(xhnhieu(arr))
```

```python
#bai6
def lamphang(lst):
    kq = []
    for i in lst:
        for items in i:
            kq.append(items)
    return kq
lst = eval(input())
print(lamphang(lst))
```

```python
#bai7
num = list(map(int, input().split()))
n = len(num)
dp = [1]*n
for i in range(n):
    for j in range(i):
        if num[j] < num[i]:
            dp[i] = max(dp[i], dp[j] + 1)
print(max(dp))
```

```python
#bai9
import bisect
#dem so vi tri co the chen
def count_smaller(nums):
    sorted_list = []
    result = []

    for num in reversed(nums):
        pos = bisect.bisect_left(sorted_list, num)
        result.append(pos)
        sorted_list.insert(pos, num)

    return result[::-1]
nums = list(map(int, input().split()))
print(count_smaller(nums))
```

```python
#bai10
def max_subarray_sum_mod_m(nums, m):
    ans = 0
    n = len(nums)

    for i in range(n):
        cur_sum = 0
        for j in range(i, n):
            cur_sum += nums[j]
            ans = max(ans, cur_sum % m)
    return ans
nums = list(map(int, input().split()))
m = int(input())
print(max_subarray_sum_mod_m(nums, m))
```
