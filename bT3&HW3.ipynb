```python
#bai16
n = int(input("Nhập số nguyên dương n: "))

chan = 0
le = 0

while n > 0:
    s = n % 10         # Lấy chữ số cuối
    if s % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10 # xong 1 vong thi bo so cuoi(la phan sau dau phay)
print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```
