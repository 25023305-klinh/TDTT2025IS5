```python
#bai 1

def reverse(path: str) -> int:
    with open(path, 'r', encoding='utf-8') as f:
        n = f.read().strip()   
    reversed_n = n[::-1]       
    return int(reversed_n)     
    
path = input().strip()     # nhập đường dẫn tới tệp
print(reverse(path))
```

```python
#bai4
import string
def normalize(s: str) -> str:
    s = s.lower()
    return ''.join(ch for ch in s if ch.isalpha())
def areAnagrams(path1: str, path2: str) -> bool:
    with open(path1, "r", encoding="utf-8") as f1:
        s1 = f1.read()
    with open(path2, "r", encoding="utf-8") as f2:
        s2 = f2.read()
    s1 = normalize(s1)
    s2 = normalize(s2)
    return sorted(s1) == sorted(s2)
with open("file1.txt", "w", encoding="utf-8") as f:
    f.write("Tom Marvolo Riddle")
with open("file2.txt", "w", encoding="utf-8") as f:
    f.write("I Am Lord Voldemort.")
print(areAnagrams("file1.txt", "file2.txt"))
```

```python
#bai5
def count(path: str, character: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        sentence = f.read()
    return sentence.count(character)
with open("text.txt", "w", encoding="utf-8") as f:
    f.write("We accept the love we think we deserve.")
print(count("text.txt", "o"))
```

```python
#bai6
def countCharacter(path: str, letter: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    return text.lower().count(letter.lower())
with open("sample.txt", "w", encoding="utf-8") as f:
    f.write("Everyone deserves a second shot.")
print(countCharacter("sample.txt", "e"))
```

```python
#bai10
def getMoney(path: str) -> list[int]:
    with open(path, "r", encoding="utf-8") as f:
        data = f.read().strip()
    if not data:
        return []
    return list(map(int, data.split()))
def rob(houses: list[int]) -> int:
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]
    prev2 = houses[0]
    prev1 = max(houses[0], houses[1])
    for i in range(2, len(houses)):
        current = max(prev1, prev2 + houses[i])
        prev2 = prev1
        prev1 = current
    return prev1
with open("money.txt", "w", encoding="utf-8") as f:
    f.write("2 7 9 3 1")
houses = getMoney("money.txt")
result = rob(houses)
print(result)
```

```python
#bai12
def maximumProduct(path: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        n = int(f.readline().strip())
        arr = list(map(int, f.readline().split()))
    arr.sort()
    product1 = arr[-1] * arr[-2] * arr[-3]
    product2 = arr[0] * arr[1] * arr[-1]
    return max(product1, product2)
with open("numbers.txt", "w", encoding="utf-8") as f:
    f.write(
        "5\n"
        "-10 -10 5 2 8")
print(maximumProduct("numbers.txt"))
```

```python
#bai13
def moveZeroes(fileName: str) -> list:
    with open(fileName, "r", encoding="utf-8") as f:
        n = int(f.readline().strip())
        arr = list(map(int, f.readline().split()))
    result = []
    zero_count = 0
    for x in arr:
        if x == 0:
            zero_count += 1
        else:
            result.append(x)
    result.extend([0] * zero_count)
    return result
with open("input.txt", "w", encoding="utf-8") as f:
    f.write(
        "7\n"
        "0 1 0 3 12 0 5")
res = moveZeroes("input.txt")
print(res)
```

```python
#bai14
def averageTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()
    total_time = 0.0
    count = 0
    for line in lines[1:]:
        parts = line.strip().split(",")
        if len(parts) >= 2:
            total_time += float(parts[1])
            count += 1
    if count == 0:
        return 0.00
    return round(total_time / count, 2)
with open("laps.csv", "w", encoding="utf-8") as f:
    f.write(
        "Lap,Time\n"
        "1,143.23\n"
        "2,134.12\n"
        "3,140.50\n")
result = averageTime("laps.csv")
print(result)
```

```python
#bai15
def findMovies(path: str):
    result = []
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()
    for line in lines[1:]:
        parts = [p.strip() for p in line.split(",")]
        movie = parts[1]
        description = parts[2].lower()
        rating = float(parts[3])
        if rating > 8.0 and description != "boring":
            result.append(movie)
    if not result:
        return "All are boring!"
    return result
with open("movies.csv", "w", encoding="utf-8") as f:
    f.write(
        "id, movie, description, rating\n"
        "1, Thor, great 3D, 8.0\n"
        "2, The Rings, Fantasy, 7.5\n"
        "3, Irish, boring, 6.2\n"
        "4, Ice song, Fantasy, 8.6\n"
        "5, Home Alone, Interesting, 6.5\n")
result = findMovies("movies.csv")
print(result)
```

```python
#bai16
import math
with open("shapes.txt", "w", encoding="utf-8") as f:
    f.write(
        "TYPE: SQUARE\n"
        "SIDE: 5.3\n"
        "\n"
        "TYPE: RECTANGLE\n"
        "WIDTH: 5.0 HEIGHT: 16.0\n"
        "\n"
        "TYPE: CIRCLE\n"
        "RADIUS: 2.15\n")
path = "shapes.txt"
with open(path, "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f.readlines()]

i = 0
while i < len(lines):
    if lines[i].startswith("TYPE:"):
        shape_type = lines[i].split(":")[1].strip()
        info = lines[i + 1]

        if shape_type == "SQUARE":
            side = float(info.split(":")[1].strip())
            perimeter = 4 * side

        elif shape_type == "RECTANGLE":
            parts = info.split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)

        elif shape_type == "CIRCLE":
            radius = float(info.split(":")[1].strip())
            perimeter = 2 * math.pi * radius

        print(f"{perimeter:.2f}")
        i += 2
    else:
        i += 1
```

```python
#bai17
def productExceptSelf(path: str) -> list[int]:
    with open(path, "r", encoding="utf-8") as f:
        arr = list(map(int, f.read().split()))
    n = len(arr)
    result = [1] * n
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= arr[i]
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= arr[i]
    return result
with open("numbers.txt", "w", encoding="utf-8") as f:
    f.write("4 6 2 5 2")
res = productExceptSelf("numbers.txt")
print(*res)
```

```python
#bai18
import math
def estimatedTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()
    src = lines[0].split()
    x1 = float(src[2])
    y1 = float(src[4])
    des = lines[1].split()
    x2 = float(des[2])
    y2 = float(des[4])
    velocity = float(lines[2].split()[1])
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    time = distance / velocity
    return round(time, 2)
with open("missile.txt", "w", encoding="utf-8") as f:
    f.write(
        "SRC X: 0 Y: 0\n"
        "DES X: 10000 Y: 10000\n"
        "VELOCITY: 6 KM/S\n")
result = estimatedTime("missile.txt")
print(result)
```

```python
#bai19
import struct
with open("data.bin", "wb") as f:
    numbers = [3.1415926535, 2.718281828, 1.23456789, 9.87654321]
    for num in numbers:
        f.write(struct.pack("d", num))  
path = "data.bin"  
with open(path, "rb") as f:
    while True:
        data = f.read(8) 
        if not data:
            break
        value = struct.unpack("d", data)[0]
        print(f"{value:.5f}")
```
