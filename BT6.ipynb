```python
#6a10
n = list(map(int, input().split()))
k = int(input())
strack = []
a = len(n)
for i in range(a):
    for j in range(i+1, a):
        if n[i] + n[j] == k:
            strack.append((n[i], n[j]))
strack.sort() #neumuon sap xep theo ptu thu i thi dung sort(key=lambda n: n[i-1])
print(strack)
            
```

```python
#6a12
n = list(map(int , input().split()))
dem = {}
for i in n:
    if i in dem:
        dem[i] += 1
    else:
        dem[i] = 1
max_dem = max(dem.values())
ans = None
for x in dem:
    if dem[x] == max_dem:
        if ans is None or x < ans:
            ans = x
print(ans)
```
