```python
#6a17
def majority_element(arr):
    candidate = None
    count = 0
    for x in arr:
        if count == 0:
            candidate = x
            count = 1
        elif x == candidate:
            count += 1
        else:
            count -= 1
    if candidate is not None and arr.count(candidate) > len(arr) // 2:
        print(candidate)
arr = list(map(int, input().split()))
majority_element(arr)
```

```python
#6a23
def dem_ten(s):
    count = 0
    word = ""
    for ch in s + " ":
        if ch.isalpha():
            word += ch
        else:
            if len(word) >= 2 and word[0].isupper() and word[1:].islower():
                count += 1
            word = ""
    return count
s = input()
print(dem_ten(s))
```

```python
#7a25
def so_lon_nhat_k_chu_so(s, k):
    result = ""
    start = 0
    n = len(s)
    for i in range(k):
        end = n - (k - i) + 1
        max_digit = '0'
        pos = start
        for j in range(start, end):
            if s[j] > max_digit:
                max_digit = s[j]
                pos = j
        result += max_digit
        start = pos + 1
    return result
s = input()
k = int(input())
print(so_lon_nhat_k_chu_so(s, k))
```

```python
#7a27
def ngoac_hop_le(s):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for ch in s:
        if ch in "({[":
            stack.append(ch)
        elif ch in ")}]":
            if not stack:
                return False
            top = stack.pop()
            if top != pairs[ch]:
                return False

    return len(stack) == 0
s = input()
print(ngoac_hop_le(s))
```

```python
#7a28
def la_hop_so(x):
    if x <= 3:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return True
    return False
def day_hop_so_dai_nhat(N):
    max_len = 0
    max_start = -1
    cur_len = 0
    cur_start = -1
    for i in range(1, N):
        if la_hop_so(i):
            if cur_len == 0:
                cur_start = i
            cur_len += 1
        else:
            if cur_len > max_len:
                max_len = cur_len
                max_start = cur_start
            cur_len = 0
    if cur_len > max_len:
        max_len = cur_len
        max_start = cur_start
    return list(range(max_start, max_start + max_len))
N = int(input())
print(*day_hop_so_dai_nhat(N))
```

```python
#7a29
def day_tang_dan_lien_tiep_dai_nhat(a):
    if not a:
        return []
    max_start = 0
    max_len = 1
    cur_start = 0
    cur_len = 1
    for i in range(1, len(a)):
        if a[i] > a[i - 1]:
            cur_len += 1
        else:
            if cur_len > max_len:
                max_len = cur_len
                max_start = cur_start
            cur_start = i
            cur_len = 1
    if cur_len > max_len:
        max_len = cur_len
        max_start = cur_start
    return a[max_start:max_start + max_len]
a = list(map(int, input().split()))
print(*day_tang_dan_lien_tiep_dai_nhat(a))
```
