```python
#bai5
#ax^2+bx+c=0
a, b , c = map(int, input().split())
denta = b*b - 4*a*c
if denta == 0:
    x = -b/2*a
    print("pt co nghiem kep x =", x)
else:
    if denta < 0:
        print("phuong trinh vo nghiem")
    if denta > 0:
        x1 = (-b + denta**0.5)/2*a
        x2 = (-b - denta**0.5)/2*a
        print("pt co 2 nghiem:", x1, x2)
```

```python
#bai7
a, b, m = map(int, input().split())
c, d, n = map(int, input().split())
det = a*d - c*b
detx = d*m - b*n
dety = a*n - c*m
if det == 0:
    if detx == dety == 0:
        print("phuong trinh vo so nghiem")
    else:
        print("phuong trinh vo nghiem")
else:
    x = detx/det
    y = dety/det
print(x, y)
```
