```python
#6a19
def vitri_max(lst):
    maxx = lst[0]
    vi_tri = 0
    for i in range(1, len(lst)):
        if lst[i] > maxx:
            maxx = lst[i]
            vi_tri = i
    return vi_tri

lst = list(map(int, input().split()))
print(vitri_max(lst))
```

```python
#6a21
n = input().split()
dic = {}
for i in range(len(n)):
    if n[i] not in dic:
        dic[n[i]] = []
    dic[n[i]].append(i)
for k in dic:
    dic[k] = tuple(dic[k])
print(dic)
    
```

```python
#6a24
def dayso(n):
    zero = [ i for i in n if i == 0]
    other = [ i for i in n if i != 0]
    return zero + other
n = list(map(int, input().split()))
print(dayso(n))
```

```python
#6a26
def tong(n):
    s = 0
    for i in range(1, len(n) - 1):
        if n[i] > n[i-1] and n[i] > n[i+1]:
            s += n[i]
    return s
n = list(map(int, input().split()))
print(tong(n))
```

```python
#6a27
def solonhon(n, x):
    check = []
    for i in range(len(n)):
        if n[i] < x:
            check.append(i)
    return check if check else [-1]

n = list(map(int, input().split()))
x = int(input())
print(solonhon(n, x))
```

```python
#6a28
def check_sequence(arr):
    if len(arr) < 2:
        return "vo danh"

    inc = True
    dec = True

    for i in range(len(arr) - 1):
        if arr[i] >= arr[i + 1]:
            inc = False
        if arr[i] <= arr[i + 1]:
            dec = False

    if inc:
        return "tang thuc su"
    if dec:
        return "giam thuc su"
    return "vo danh"
arr = list(map(int, input().split()))
print(check_sequence(arr))
```
